```python
def down_block(x, filters, kernel_size=(3, 3), padding="same", strides=1):
    c = keras.layers.Conv2D(filters, kernel_size, padding=padding, strides=strides, activation="relu")(x)
    c = keras.layers.Conv2D(filters, kernel_size, padding=padding, strides=strides, activation="relu")(c)
    p = keras.layers.MaxPool2D((2, 2), (2, 2))(c)
    return c, p

def up_block(x, skip, filters, kernel_size=(3, 3), padding="same", strides=1):
    us = keras.layers.UpSampling2D((2, 2))(x)
    concat = keras.layers.Concatenate()([us, skip])
    c = keras.layers.Conv2D(filters, kernel_size, padding=padding, strides=strides, activation="relu")(concat)
    c = keras.layers.Conv2D(filters, kernel_size, padding=padding, strides=strides, activation="relu")(c)
    return c

def bottleneck(x, filters, kernel_size=(3, 3), padding="same", strides=1):
    c = keras.layers.Conv2D(filters, kernel_size, padding=padding, strides=strides, activation="relu")(x)
    c = keras.layers.Conv2D(filters, kernel_size, padding=padding, strides=strides, activation="relu")(c)
    return c

def UNet():
    f = [4,8,16,32]
    inputs  = keras.layers.Input((64,64,43))
    
    p0 = inputs
    c1, p1 = down_block(p0, f[0])#64 -> 32
    c2, p2 = down_block(p1, f[1]) #32 -> 16
    c3, p3 = down_block(p2, f[2]) #16->8
    
    bn = bottleneck(p3, f[3])
    
    u2 = up_block(bn, c3, f[2]) #8  -> 16
    u3 = up_block(u2, c2, f[1]) #16 -> 32
    u4 = up_block(u3, c1, f[0]) #32 -> 64
    
    outputs = keras.layers.Conv2D(1, (1, 1), padding="same", activation="sigmoid")(u4)
    model = keras.models.Model(inputs, outputs)
    return model
```

```python
import tensorflow.keras as keras
from sklearn import model_selection
from tensorflow.keras import backend
# force channels-first ordering
backend.set_image_data_format('channels_last')
print(backend.image_data_format())

checkpoint_path = "training_1/cp.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)
images = pickle.load(open( 'good_img', "rb" ))
labels = pickle.load(open( 'good_label_binarized', "rb" ))

# labels = np.array([train_image for train_image in labels_good])


##train test split, let's think about K-fold
train_images, test_images, train_labels, test_labels = model_selection.train_test_split(images,
                                                    labels,
                                                    test_size=0.2,
                                                    random_state=42)


# train_images = np.array([train_image for train_image in train_images])
# print(train_images.shape)
# train_images = np.swapaxes(train_images,1,3)
# train_labels = np.array([train_image for train_image in train_labels])
# print(train_labels.shape)

# test_images = np.array([train_image for train_image in test_images])
# test_images = np.swapaxes(test_images,1,3)
# test_labels = np.array([train_image for train_image in test_labels])


## training process
cp_callback = tf.keras.callbacks.ModelCheckpoint(checkpoint_path, 
                                                 save_weights_only=True,
                                                 verbose=1,
                                                 period =3)
print(len(train_images))
model = UNet()
model.compile(optimizer='adam', 
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.summary()
history = model.fit(train_images, train_labels, epochs=20,validation_data = (test_images,test_labels),
          callbacks = [cp_callback])
plt.figure()
plt.plot(history.history['loss'])
plt.show()
loss, acc = model.evaluate(test_images, test_labels)
print("Trained model, accuracy: {:5.2f}%".format(100*acc))
```
